# Handle Missing Categorical Data

**1.Frequent Category Imputation**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load Titanic dataset
df = sns.load_dataset('titanic')

# Check missing values in 'embark_town' column
print("Missing values before imputation:")
print(df['embark_town'].isnull().sum())

# Impute with the most frequent category
mode_value = df['embark_town'].mode()[0]
df['embark_town_frequent'] = df['embark_town'].fillna(mode_value)

# Check missing values after imputation
print("\nMissing values after frequent category imputation:")
print(df['embark_town_frequent'].isnull().sum())

# Plot before and after
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.countplot(x='embark_town', data=df)
plt.title("Before Imputation")

plt.subplot(1, 2, 2)
sns.countplot(x='embark_town_frequent', data=df)
plt.title("After Frequent Category Imputation")
plt.show()
```

**2. New Category Imputation**

```python
# Impute with a new category 'Unknown'
df['embark_town_new_category'] = df['embark_town'].fillna("Unknown")

# Plot before and after
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.countplot(x='embark_town', data=df)
plt.title("Before Imputation")

plt.subplot(1, 2, 2)
sns.countplot(x='embark_town_new_category', data=df)
plt.title("After New Category Imputation ('Unknown')")
plt.show()
```
