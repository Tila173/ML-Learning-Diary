## 1. Bivariate Analysis

Bivariate analysis involves analyzing the relationship between two variables. This can include:
* Numerical vs. Numerical: Correlation, scatter plots
* Categorical vs. Numerical: Box plots, violin plots
* Categorical vs. Categorical: Cross-tabulation, bar plots, count plots

### a. Scatter Plot (Numerical vs. Numerical)

```python
sns.scatterplot(df, x='total_bill', y='tip') #Bivariate
#sns.scatterplot(df, x='total_bill', y='tip', hue=df['sex'], style=df['smoker'],size=df['size']) #Multivariate
plt.title('Total Bill vs. Tip')
plt.show()
```

### b. Correlation Heatmap (Numerical vs. Numerical)

```python
# Select only numeric columns for the correlation matrix
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Plot the correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

### c. Box Plot (Categorical vs. Numerical)

```python
# Example: Survival by Gender
sns.boxplot(x='Sex', y='Age', data=titanic) #Bivariate
plt.title('Survival Count by Gender')
plt.show()
```

### d. Bar Plot (Categorical vs. Numerical)

```python
# Plot a bar plot with Pclass on the x-axis and Fare on the y-axis
sns.barplot(x='Pclass', y='Fare', data=titanic) #Bivariate
#sns.barplot(x='Pclass', y='Fare', data=titanic, hue=titanic['Sex']) # Multivariate

plt.title('Average Age by Passenger Class')
plt.show()
```

### e. Distplot (Categorical-Numerical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

### f. HeatMap Plot (Categorical vs. Categorical)

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

### g. ClusterMap (Categorical vs. Categorical)

```python
sns.clustermap(pd.crosstab(titanic['SibSp'], titanic['Survived']))
```

### h. Pairplot (Categorical vs. Categorical)

```python
sns.pairplot(titanic, hue='Survived', palette='coolwarm', diag_kind='kde')

plt.title('Pairplot of Titanic Dataset')
plt.show()
```

```python
# Create a pairplot with hue based on species
sns.pairplot(iris, hue='species', palette='coolwarm', diag_kind='kde')

plt.title('Pairplot of Iris Dataset')
plt.show()
```
