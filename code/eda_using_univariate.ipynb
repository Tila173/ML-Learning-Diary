## 1. Categorical Data

#### a. Countplot

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Check if 'Survived' column exists and has data
if 'Survived' in df.columns:
    ax = sns.countplot(x=df['Survived'])
    
    total = len(df['Survived'])
    for p in ax.patches:
        height = p.get_height()
        percentage = (height / total) * 100
        ax.annotate(f'{percentage:.2f}%', 
                    (p.get_x() + p.get_width() / 2., height / 2.), 
                    ha='center', va='center', 
                    fontsize=12, color='white') 
    
    plt.title("Survival Count")
    plt.xlabel('Survived')
    plt.ylabel('Count')
    
    # Show the plot
    plt.show()
else:
    print("Column 'Survived' not found.")
```

#### b. PieChart

```python
# Check if 'Survived' column exists and has data
if 'Survived' in df.columns:
    # Calculate counts for each category (0: Not Survived, 1: Survived)
    survived_counts = df['Survived'].value_counts()
    
    # Define labels and sizes for the pie chart
    labels = ['Not Survived', 'Survived']
    sizes = [survived_counts.get(0, 0), survived_counts.get(1, 0)] 
    
    # Create a pie chart
    plt.figure(figsize=(6, 6))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['yellow', 'white'])
    
    # Add title
    plt.title("Survival Distribution")
    
    # Show the plot
    plt.show()
else:
    print("Column 'Survived' not found.")
```

## 2. Numerical Data

#### a. Histogram

```python
# Check if 'Age' column exists and has data
if 'Age' in df.columns:
    plt.figure(figsize=(8, 6)) 
    
    # Create the histogram with custom styling
    plt.hist(df['Age'].dropna(), bins=30, color='skyblue', edgecolor='black', alpha=0.7)
    
    # Add titles and labels with better font styling
    plt.title('Distribution of Age', fontsize=16, fontweight='bold')
    plt.xlabel('Age', fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    
    # Add grid for better readability
    plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
    
    # Display the plot
    plt.tight_layout()
    plt.show()

else:
    print("Column 'Age' not found.")
```

#### b. Distplot

```python
# Check if 'Age' column exists and has data
if 'Age' in df.columns:
    plt.figure(figsize=(8, 6)) 

    # Create a beautiful distplot with seaborn
    sns.distplot(df['Age'].dropna(), bins=30, kde=True, color='skyblue', 
                 hist_kws={"edgecolor": "black"}, kde_kws={"color": "orange", "lw": 2})
    
    # Add titles and labels with better font styling
    plt.title('Age Distribution with Kernel Density Estimate', fontsize=16, fontweight='bold')
    plt.xlabel('Age', fontsize=14)
    plt.ylabel('Density / Frequency', fontsize=14)
    
    # Customize gridlines for aesthetics
    plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
    
    # Display the plot
    plt.tight_layout()
    plt.show()

else:
    print("Column 'Age' not found.")
```

#### c. Boxplot

```python
# Check if 'Fare' column exists and has data
if 'Fare' in df.columns:
    plt.figure(figsize=(8, 6))  # Set figure size

    # Create a box plot for 'Fare'
    sns.boxplot(x=df['Fare'], color='lightblue', fliersize=5)

    # Add titles and labels with better font styling
    plt.title('Fare Distribution Boxplot', fontsize=16, fontweight='bold')
    plt.xlabel('Fare', fontsize=14)
    
    # Customize gridlines and ticks for aesthetics
    plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
    
    # Display the plot
    plt.tight_layout()
    plt.show()

else:
    print("Column 'Fare' not found.")
```
