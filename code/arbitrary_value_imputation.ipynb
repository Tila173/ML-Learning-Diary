# Arbitrary Value Imputation

**1. Using Pandas for Arbitrary Imputation**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load Titanic dataset
df = sns.load_dataset('titanic')

# Visualize missing values in the 'age' column
plt.figure(figsize=(8, 4))
sns.histplot(df['age'], kde=True)
plt.title("Age Distribution Before Imputation")
plt.show()

# Impute missing values in 'age' column with arbitrary value -1
df['age_arbitrary'] = df['age'].fillna(-1)

# Visualize after imputation
plt.figure(figsize=(8, 4))
sns.histplot(df['age_arbitrary'], kde=True)
plt.title("Age Distribution After Arbitrary Value Imputation (-1)")
plt.show()
```

**2. Using Scikit-Learns SimpleImputer for Arbitrary Imputation**

```python
from sklearn.impute import SimpleImputer

# Initialize SimpleImputer with a specified arbitrary value
imputer = SimpleImputer(strategy='constant', fill_value=-1)

# Fit and transform 'age' column
df['age_arbitrary_sklearn'] = imputer.fit_transform(df[['age']])

# Visualize after imputation with Scikit-Learn
plt.figure(figsize=(8, 4))
sns.histplot(df['age_arbitrary_sklearn'], kde=True)
plt.title("Age Distribution After Arbitrary Value Imputation with Scikit-Learn (-1)")
plt.show()
```
