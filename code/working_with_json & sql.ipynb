### Working with SQL

If you are working with larger databases like MySQL or PostgreSQL, you can use Python libraries like mysql-connector-python for MySQL or psycopg2 for PostgreSQL.

**MySQL Example**

You can connect to a MySQL database, run queries, and even load SQL files.
* !pip install mysql-connector-python

```python
import mysql.connector

# Connect to MySQL Database
conn = mysql.connector.connect(
    host='localhost',
    user='your_username',
    password='your_password',
    database='your_database'
)

cursor = conn.cursor()

# Read and Execute SQL File
with open('database.sql', 'r') as sql_file:
    sql_script = sql_file.read()

# Execute the SQL script
for statement in sql_script.split(';'):
    if statement.strip():  
        cursor.execute(statement)

# Query the Database
cursor.execute("SELECT * FROM your_table")
rows = cursor.fetchall()

for row in rows:
    print(row)

# Commit and Close the Connection
conn.commit()
conn.close()
```
